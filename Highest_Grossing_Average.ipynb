```python
import spacy
import sys
sys.path.append('tomatopy-master')
sys.path.append('tomatopy-master/tomatopy')
from wikipedia import scrape_movie_names
from scraper import scrape_movie_score
import tomatopy as rtp
import datetime
from matplotlib import pyplot as plt
import matplotlib.patches as mpatches
```

```python
movie_list = scrape_movie_names(2001)
print(movie_list)
```

```python
year = datetime.datetime.today().year
YEARS = [year - i for i in range(60)]
```

```python
print(YEARS)
```

```python
average_critic = []
average_audience = []
for year in YEARS:
    movie_list = scrape_movie_names(year)
    critic_sum = 0
    critic_divisible = 10
    audience_sum = 0
    audience_divisible = 10
    crit = 0
    aud = 0
    for movie in movie_list:
        try:
            C_score = scrape_movie_score(movie, 'critic')
            critic_sum += int(C_score)
            crit += 1
        except:
            critic_divisible -= 1
        try:
            A_score = scrape_movie_score(movie, 'audience')
            audience_sum += int(A_score)
            aud += 1
        except:
            audience_divisible -= 1
            continue
            
    C_ave = critic_sum/critic_divisible
    A_ave = audience_sum/audience_divisible
    average_critic.append([year, C_ave])
    average_audience.append([year, A_ave])
    print("Done with " + str(year) + ": Critic score was " + str(C_ave) + ", Audience score was " + str(A_ave))
```

```python
print(average_critic)
```

```python
print(average_audience)
```

```python
years = []
scores = []
for val in average_critic:
    years.append(val[0])
    scores.append(val[1])
plt.plot(years, scores)
```

```python
years_aud = []
scores_aud = []
for val in average_audience:
    years_aud.append(val[0])
    scores_aud.append(val[1])
plt.plot(years, scores, 'r')
plt.plot(years_aud, scores_aud, 'b')
red_patch = mpatches.Patch(color='red', label='Critic Score')
blue_patch = mpatches.Patch(color='blue', label='Audience Score')
plt.legend(handles=[red_patch, blue_patch])
plt.title("Average Rotten Tomatoes Score of Top 10 Highest Grossing Films Per Year")
plt.xlabel("")
plt.show()
```
